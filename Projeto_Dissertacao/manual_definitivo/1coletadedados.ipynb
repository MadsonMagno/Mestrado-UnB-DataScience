```python
import requests

def check_subdomains(file_path):
    ativos = []
    inativos = []

    with open(file_path, 'r') as file:
        subdomains = file.read().splitlines()

    for subdomain in subdomains:
        url = f"https://{subdomain}"
        try:
            response = requests.get(url, timeout=5)
            if response.status_code == 200:
                ativos.append(subdomain)
            else:
                inativos.append(subdomain)
        except requests.RequestException:
            inativos.append(subdomain)

    return ativos, inativos



```

```python

# chama funcionalidade
arquivo = 'sublist3r.txt'
ativos, inativos = check_subdomains(arquivo)
```

```python

# exibir os resultados
print("Subdomínios Ativos:")
print("\n".join(ativos))

print("\nSubdomínios Inativos:")
print("\n".join(inativos))
```

```python
# salva ativos no arquivo url_ativos.txt
with open('url_ativos.txt', 'w') as file:
    for subdomain in ativos:
        file.write(subdomain + '\n')
```

```python
# compara urls do arquivo url_ativos_dnsdumpster.txt estão no arquivo url_ativos.txt
with open('url_ativos_dnsdumpster.txt', 'r') as file:
    ativos_dnsdumpster = set(file.read().splitlines())
with open('url_ativos.txt', 'r') as file:
    ativos = set(file.read().splitlines())
# encontra urls que estão em ambos os arquivos
comum = ativos_dnsdumpster.intersection(ativos)
# exibe as urls que estão em ambos os arquivos
print("Subdomínios comuns em ambos os arquivos:")
print("\n".join(comum))
```

```python
# salva subdominios comuns em um arquivo comum.txt
with open('comum.txt', 'w') as file:
    for subdomain in comum:
        file.write(subdomain + '\n')    
```
